```python
# Con base en la metadata, determinamos cuales son variables categricas
columnas_categ = ['X2','X3', 'X4', 'X5']
```

```python
# Convertimos variables numricas a categricas como corresponde (cambio de tipo de dato)
df_cat = df.copy()
for col in columnas_categ:
  df_cat[col] = df_cat[col].astype('category')
df_cat
```
